```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
import datetime as dt
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM
```

```python
import yfinance as yf
```

```python
indian_stocks = ['RELIANCE.NS', 'TCS.NS', 'INFY.NS', 'HDFCBANK.NS', 'ICICIBANK.NS']
```

```python
start = dt.datetime(2012, 1, 1)
end = dt.datetime(2023, 1, 1)
```

```python
for stock in indian_stocks:
    print(f"Processing {stock}...")
    try:
        # Load the data using yfinance
        data = yf.download(stock, start=start, end=end)
        
        if data.shape[0] == 0:
            print(f"No data found for {stock}. Skipping...")
            continue
        
        # Prepare the data for LSTM model
        scaler = MinMaxScaler(feature_range=(0, 1))
        scaled_data = scaler.fit_transform(data['Close'].values.reshape(-1, 1))
        
        prediction_days = 60
        
        x_train = []
        y_train = []
        
        for x in range(prediction_days, int(len(scaled_data) * 0.8)):
            x_train.append(scaled_data[x-prediction_days:x, 0])
            y_train.append(scaled_data[x, 0])
        
        x_train, y_train = np.array(x_train), np.array(y_train)
        x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
        
        # Build the LSTM model
        model = Sequential()

        model.add(LSTM(units=50, return_sequences=True, input_shape=(x_train.shape[1], 1)))
        model.add(Dropout(0.2))
        model.add(LSTM(units=50, return_sequences=True))
        model.add(Dropout(0.2))
        model.add(LSTM(units=50))
        model.add(Dropout(0.2))
        model.add(Dense(units=1))  # Prediction of the next closing value

        model.compile(optimizer='adam', loss='mean_squared_error')
        model.fit(x_train, y_train, epochs=25, batch_size=32)
        
        # Make training predictions
        train_predictions = model.predict(x_train)
        train_predictions = scaler.inverse_transform(train_predictions)
        
        # Prepare validation data
        val_data = scaled_data[int(len(scaled_data) * 0.8) - prediction_days:]
        x_val = []
        y_val = []
        for x in range(prediction_days, len(val_data)):
            x_val.append(val_data[x-prediction_days:x, 0])
            y_val.append(val_data[x, 0])

        x_val, y_val = np.array(x_val), np.array(y_val)
        x_val = np.reshape(x_val, (x_val.shape[0], x_val.shape[1], 1))
        
        # Make validation predictions
        val_predictions = model.predict(x_val)
        val_predictions = scaler.inverse_transform(val_predictions)
        
        # Test the model accuracy on existing data
        test_start = dt.datetime(2023, 1, 1)
        test_end = dt.datetime.now()

        test_data = yf.download(stock, start=test_start, end=test_end)
        
        if test_data.shape[0] == 0:
            print(f"No test data found for {stock}. Skipping...")
            continue
        
        actual_prices = test_data['Close'].values
        
        total_dataset = pd.concat((data['Close'], test_data['Close']), axis=0)

        model_inputs = total_dataset[len(total_dataset) - len(test_data) - prediction_days:].values
        model_inputs = model_inputs.reshape(-1, 1)
        model_inputs = scaler.transform(model_inputs)

        # Make predictions on test data
        x_test = []

        for x in range(prediction_days, len(model_inputs)):
            x_test.append(model_inputs[x-prediction_days:x, 0])
        
        x_test = np.array(x_test)
        
        if x_test.size == 0:
            print(f"Not enough data to reshape for {stock}. Skipping...")
            continue
        
        x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))
        
        predicted_prices = model.predict(x_test)
        predicted_prices = scaler.inverse_transform(predicted_prices)
        
        # Recursive predictions
        last_train_input = x_test[-1]
        recursive_predictions = []
        for _ in range(60):
            pred = model.predict(last_train_input.reshape(1, -1, 1))
            recursive_predictions.append(pred[0, 0])
            last_train_input = np.append(last_train_input[1:], pred, axis=0)
        
        recursive_predictions = scaler.inverse_transform(np.array(recursive_predictions).reshape(-1, 1))
        
        # Plot the results
        plt.figure()  # Ensure separate figures for each stock
        plt.plot(np.arange(prediction_days, len(train_predictions) + prediction_days), train_predictions, color="yellow", label="Training Predictions")
        plt.plot(np.arange(prediction_days, len(y_train) + prediction_days), scaler.inverse_transform(y_train.reshape(-1, 1)), color="blue", label="Training Observations")
        plt.plot(np.arange(len(train_predictions) + prediction_days, len(train_predictions) + prediction_days + len(val_predictions)), val_predictions, color="green", label="Validation Predictions")
        plt.plot(np.arange(len(train_predictions) + prediction_days, len(train_predictions) + prediction_days + len(val_predictions)), scaler.inverse_transform(y_val.reshape(-1, 1)), color="lightgreen", label="Validation Observations")
        plt.plot(np.arange(len(total_dataset) - len(test_data), len(total_dataset)), predicted_prices, color="purple", label="Testing Predictions")
        plt.plot(np.arange(len(total_dataset), len(total_dataset) + len(recursive_predictions)), recursive_predictions, color="red", label="Recursive Predictions")
        plt.title(f"{stock} Share Price Predictions")
        plt.xlabel('Time')
        plt.ylabel(f'{stock} Share Price')
        plt.legend()
        plt.show()
    
    except Exception as e:
        print(f"Failed to process {stock}. Error: {e}")
```
